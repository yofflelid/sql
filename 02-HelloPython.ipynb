```python
# 비교식을 사용한 if 문

myage = int(input('나이를 입력하세요.'))

if myage == 19:
    print("19살 입니다.")

if myage != 20:
    print("20살이 아닙니다.")

if myage > 19:
    print("성인입니다.")

if myage <= 19:
    print("청소년 입니다.")
```

```python
# 논리식을 사용한 if 문


point = int(input("점수를 입력하세요"))

#AND --> 전체가 참이어야 결과가 참 

if point > 70 and point <= 80:
    print("C학점 입니당당다다다다다다다다")

# OR --> 하나라도 참이면 전체가 참 

if point <= 70 or point > 80:
    print("C학점이 아닙니다.")
```

```python
# 논리값에 대한 조건문 

#isKorean = True
isKorean = False 

if isKorean == True:
    print("한국 사람 입니다...(1)")

if isKorean == False:
    print("한국 사람이 아닙니다...(2)")

if isKorean != True:
    print("한국 사람이 아닙니다...(3)")

if isKorean != False:
    print("한국 사람 입니다...(4)")

if isKorean:
    print("한국 사람 입니다...(5)")

if not isKorean:
    print("한국 사람이 아닙니다...(6)")
```

```python
# else문 활용하기 

# 입력값에 따른 if ~ else 분기 파악하기 

print("시험시작")

score = int(input("점수를 입력하세요"))

if score >= 60:
    print("합격 입니다.")
else: 
    print("불합격 입니다.")

print("시험 끝")
```

# 5. 반복문

```python


x = 1
while x <= 10:
    print("x=", x)
    x += 1

print("x의 최종값=", x)
```

```python
# while 문의 기본 구조 

# 1부터 10까지 1씩 증가하는 조건에서의 반복문 

#처음에 값을 셋팅을 해야겠지? 
x = 100
while x <= 500:
    print( "x =", x)

    x += 100
    
print("x의 최종값으으으은 =", x)
```

```python
x = 100
while x <= 500:
    print( "x =", x)

    x += 100
    
    print("x의 최종값으으으은 =", x)
```

```python
x = int(input("숫자를 입력하세요."))

# 구구단 출력하기 
y = 1                #초기식 
while y < 10:       #조건식 -> 10보다 작다 이므로 10은 포함되지 않음 
    z = x * y          # 반복이 진행되는 동안 수행할 명령 
    print(x, "x", y, "=", z)
    y += 1          # 증감식 -> y가 1씩 증가함 
```

```python
# 1~10 까지의 총 합을 구하기 

x = 1                       #초기식 
sum = 0                     #총 합을 누적해서 저장할 변수 
while x <= 10:              #조건식 -> x가 10보다 작거나 같은 동안 반복 수행 
    sum += x
    print("x=", x, "-> sum =", sum)
    x += 1                  #증감식 -> 조건에 사용되는 값을 변경 


# 왜 x +=1 을 프린트 밑에 적는거야? 
```

```python
# 값의 변화 단계 조절하기 

# 0부터 50전 (=49)까지 10씩 증가 

a = 0 
while a < 50:
    print(a)
    a += 10 
```

```python
# 10부터 0전(=1)까지 2씩 감소 

b = 10 
while b > 0:
    print( b )
    b -= 2
```

### for문 기본 구조

```python
# x가 1부터 11전까지 1씩 증가하면서 반복 수행 

for x in range(1,11):
    # 반복이 진행되는 동안 수행할 명령 
    print("x=", x)

print("x의 최종값=", x)
```

```python
# 구구단 출력하기 

x = int(input("숫자를 입력하세요. 정수로 !!!!!!!!!!!!!"))

# -> 1부터 10전까지 1씩 증가하는 값을 y에 설정 
for y in range(1,10):
    #반복이 진행되는 동안 수행할 명령
    z = x * y
    print( x, "x", y, "=", z)
```

```python
# 합계 구하기 

# 1~10까지의 총 합을 구하기 
sum = 0 #총 합을 누적해서 저장할 변수  

for x in range(1,11): #x가 1부터 11전까지 1씩 증가하면서 반복수행 
    #sum이 x만큼 스스로 증가 
    #-> x가 1~10까지 변화하므로 sum은 1부터 10까지 순차적으로 증가 
    sum += x
    print("x=", x, "-> sum =", sum)

```

```python
# 값의 변화 단계 조절하기 

# range() 함수의 세번째 파라미터로 변화량을 설정할 수 있다. 

#0부터 50전(=49)까지 10씩 중가 

for a in range(0, 50, 10):
    print( a )
```

```python
# 10부터 0전(=1)까지 2씩 감소 

for b in range(10, 0, -2):
    print( b )
```
